## Задание 3. Построение графиков с использованием Matplotlib

### 3.1. Построить несколько графиков разного вида с использованием matplotlib

```python
#  Диаграмма рассеяния между рейтингом и продолжительностью фильма
import matplotlib as mpl
mpl.rcParams['agg.path.chunksize'] = 10000

fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(imdb_data.rating,
           imdb_data.runtime,
           color='#1f77b4',
           alpha=0.7
        )
ax.set(title='rating and runtime', xlabel='rating', ylabel='runtime')
plt.show()
```

```python
# Ящик с усами для распределения рейтингов фильмов
plt.figure(figsize=(10, 6))
plt.boxplot(imdb_data.rating, patch_artist=True)
plt.title('Mustache Box: Film Ratings Distribution')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

### 3.2. Построить несколько суб-графиков на одном графике используя .subplot

```python
# Комбинация различных типов графиков и диаграмм, которые отображают различные характеристики
fig, ax = plt.subplots(2,2, figsize=(18, 18))
# Гистограмма оценок фильмов (metascore)
ax[0,0].hist(imdb_data.metascore, color="r")
ax[0,0].set(title = "metascore")
# Круговая диаграмма для рейтинга фильмов
ax[0, 1].pie(imdb_data.rating.value_counts(), labels=imdb_data.rating.value_counts().index, autopct='%1.1f%%')
ax[0, 1].set(title ="Rating distribution")
# Столбчатые диаграммы для распределения времени просмотра фильмов (runtime)
ax[1,0].bar(imdb_data.runtime.value_counts().index, imdb_data.runtime.value_counts().values, color="g")
ax[1,0].set(title ="runtime")
# Скрипичная диаграмма для рейтинга фильмов (rating)
ax[1, 1].violinplot(imdb_data.rating)
ax[1, 1].set(title ="rating")

plt.show()
```

### 3.3. Построить интерактивный график со слайдером.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interact, IntSlider

# Функция для построения интерактивного Distribution Plot
def plot_rating_distribution(year):
    """
    Построение графика распределения рейтингов фильмов.
    Аргументы:
    year (int): Год, для которого строится распределение рейтингов.
    """
    # Фильтруем датасет по указанному году
    selected_year_data = imdb_data[imdb_data.year == year]

    # Строим Distribution Plot
    plt.figure(figsize=(10, 6))
    sns.histplot(selected_year_data.rating, kde=True, color='purple')
    plt.title('Distribution of film ratings for a given year')
    plt.xlabel('Raiting')
    plt.ylabel('Number of films')
    plt.grid(True)
    plt.show()

# Создаем слайдер для выбора года
year_slider = IntSlider(min=imdb_data.year.min(), max=imdb_data.year.max(), step=1, value=imdb_data.year.min(), description='Year of release')

# Используем функцию interact для создания интерактивного графика
interact(plot_rating_distribution, year=year_slider)
```

### 5.3. Построить .boxplot и/или .violinplot на каком-либо наборе данных, проанализировать числовой признак в разрезе категориального.

```python
imdb_data1 = imdb_data.sort_values(by=['year'])
sns.violinplot(data=imdb_data1,
               x='year',  # x присвоен параметру hue
               y='rating',
               hue='year',  # новый параметр hue
               palette="Set2",
               bw_method=0.2,  # новый параметр bw_method
               cut=1,
               linewidth=3,
               legend=False)  # установка legend=False для предотвращения отображения легенды
plt.xlim(80, 95)
plt.ylim(7, 9)
plt.show()
```

### 5.4. Построить график .heatmap попарной корреляции всех числовых признаков.

```python
# Выбираем только числовые признаки для вычисления корреляции
numeric_features = imdb_data.select_dtypes(include=['int64', 'float64'])

# Вычисляем корреляцию методом Спирмена
corr = numeric_features.corr(method='spearman')

# Создаем маску для скрытия верхнего треугольника корреляционной матрицы
mask = np.triu(np.ones_like(corr, dtype=bool))

# Создаем объект subplot для тепловой карты
f, ax = plt.subplots(figsize=(10, 10))

# Создаем цветовую карту для тепловой карты
cmap = sns.diverging_palette(200, 21, as_cmap=True)

# Строим тепловую карту
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

# Показываем тепловую карту
plt.show()
```

### 7.2. Построить несколько суб-графиков на одном графике.

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Создаем объект subplots
fig = make_subplots(rows=1, cols=2)

# Добавляем - Гистограмма голосов
fig.add_trace(
    go.Histogram(x=imdb_data['votes'], name='Votes Histogram'),
    row=1, col=1
)

# Добавляем - Boxplot для переменной runtime
fig.add_trace(
    go.Box(y=imdb_data['runtime'], name='Runtime Boxplot'),
    row=1, col=2
)

# Обновляем макет графика
fig.update_layout(title='Subplots: Гистограмма голосов и Boxplot для runtime')

# Показываем график
fig.show()
```

```python
# Интерактивный scatter plot, отображающий зависимость количества голосов от времени просмотра фильма
import dash
from dash import dcc, html
import plotly.express as px
import pandas as pd

# Создание экземпляра Dash приложения
app = dash.Dash(__name__)

# Определение макета приложения
app.layout = html.Div([
    dcc.Graph(
        id='interactive-graph',
        figure=px.scatter(imdb_data, x='votes', y='runtime', title='Dependence of the number of votes on the viewing time of the film')
    )
])

# Запуск сервера Dash
if __name__ == '__main__':
    app.run_server(debug=True)
```
