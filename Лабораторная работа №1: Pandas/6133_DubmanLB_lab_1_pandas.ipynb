### 7) Создание нового поля, вычисленного на основе значений других полей.

Перед выполнением данного задания, произведем небольшую корректировку DataFrame: вещественные показатели продаж приведем к единичным экземплярам, поскольку на данный момент они исчисляются в миллионах экземпляров.

```python
data.loc[data["NA_Sales"] > 0, "NA_Sales"]=data["NA_Sales"]*1000000;
data.loc[data["EU_Sales"] > 0, "EU_Sales"]=data["EU_Sales"]*1000000;
data.loc[data["JP_Sales"] > 0, "JP_Sales"]=data["JP_Sales"]*1000000;
data.loc[data["Other_Sales"] > 0, "Other_Sales"]=data["Other_Sales"]*1000000;
data.loc[data["Global_Sales"] > 0, "Global_Sales"]=data["Global_Sales"]*1000000;
```

#### 7.1) Через выражение на базе имеющихся колонок

Создадим несколько новых колонок, но без имеющихся колонок.
На основе столбцов отображающих, количество копий проданных в различных регионах, в том числе и по всему миру,
cоздадим новые, которые бы отражали возможную прибыль при продаже такого количества копий игр.

```python
data["NA_Sales_Dollars"] = data["NA_Sales"]*np.random.randint(1, 100);
data["EU_Sales_Dollars"] = data["EU_Sales"]*np.random.randint(1, 100);
data["JP_Sales_Dollars"] = data["JP_Sales"]*np.random.randint(1, 100);
data["Other_Sales_Dollars"] = data["Other_Sales"]*np.random.randint(1, 100);
data["Global_Sales_Dollars"] = data["Global_Sales"]*np.random.randint(1, 100);
```

Посмотрим результат

```python
data
```

#### 7.2) Через DataFrame.apply

Создадим новую колонку с данными используя DataFrame.apply, которая будет говорить нам о прибыльности той или иной игры:
Игра является прибыльной в случае, если ее мировые сборы превысят 50 млн долларов США.

```python
data["Money_Status"] = data.apply(lambda x: "Profitable" if x["Global_Sales_Dollars"] > 50000000 else "Not profitable", axis=1)
```

Проверим результат

```python
data.sample(10)
```

#### 7.3) через Series.apply

Создадим новую колонку с данными используя Series.apply, которая будет говорить нам о принадлежности той или иной игры к поколению игр:
Игра является игрой нового поколения в случае, если она выпущена была после 2007 года.

```python
data["Status_Gen"] = data["Year"].apply(lambda x: "New Gen" if x > 2007 else "Old Gen")
```

Посмотрим результат

```python
data.sample(10)
```

## Задание 5. Решейпинг данных 1D -> 2D с использованием .pivot (можно подать на вход результаты агрегатов, полученных ранее с использованием .groupby (сгруппировать по двум полям).

```python
data_reshape = data.groupby(["Year", "Genre"])["Global_Sales_Dollars"].max().reset_index()
data_reshape.pivot(index="Year", columns="Genre")
data_reshape
```

В результате выполнения данной операции мы получили данные о максимальной прибыли для каждого жанра в конкретном году.

```python
data_reshape_table = data.pivot_table(index=["Year"], columns=["Genre"], values="Global_Sales_Dollars", aggfunc="max")
data_reshape_table[data_reshape_table.notnull() == True]
```

В результате выполнения данной операции мы получили данные о максимальной прибыли для каждого жанра в конкретном году.
